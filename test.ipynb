```python
import pandas as pd
import seaborn as sb
met = pd.read_csv("metabolites.csv")
means = met.groupby(['genus'], as_index = False).aggregate({'mean'})
means
sb.heatmap(means, cmap="YlGnBu", annot=True)
```

```python
my_iter = map(lambda x: x%3==0, [1,2,3,4,5,6])
while True or False:
    try:
        print(next(my_iter))
    except StopIteration:
        print ("The End of List")
        break
# Therefore iterators might be more preferable than list to catch something (like particular number or an error)
```
